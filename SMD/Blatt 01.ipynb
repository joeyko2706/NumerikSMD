Die Funktion $$f(x) = (x^3 + \frac 13) - (x^3 - \frac 13)$$ ist für $$ x\in\mathbb R$$ gleich $$\frac 13.
$$ Numerisch folgt:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,100,1000) #Einen Bereich für x von 0 bis 100 mit 1000 Knoten erstellen
y = (x**3 + 1/3) - (x**3 -  1/3)


plt.plot(x, y, color='mediumblue', label='Numerische Funktion')
plt.plot(1/3, color='crimson', label='Algebraischer Wert')
plt.grid()
plt.legend(loc='best')
plt.show()
```
